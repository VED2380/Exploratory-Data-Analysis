```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [1,4,9,16,25]

## Create a line plot
plt.plot(x,y)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Basic line plot')
plt.show()
```

```python
x = [1,2,3,4,5]
y = [1,4,9,16,25]

## Customized line plot

plt.plot(x,y,color='red',linestyle ="--",marker='o',linewidth=3,markersize=9)
plt.grid(True)
```

```python
plt.plot(x,y,color='red',linestyle ="-.",marker='*')
```

```python
## Multiple Plots
## Sample data

x = [1,2,3,4,5]
y1 =[1,4,6,16,25]
y2 = [1,2,3,4,5]

plt.figure(figsize=(9,5))
plt.subplot(1,2,1)
plt.plot(x,y1,color='green')
plt.title('Plot1')

plt.subplot(1,2,2)
plt.plot(x,y2,color='red')
plt.title('Plot2')

```

```python
plt.subplot(2,2,1)
plt.plot(x,y1,color='green')
plt.title('Plot1')

plt.subplot(2,2,2)
plt.plot(x,y2,color='red')
plt.title('Plot2')

plt.subplot(2,2,3)
plt.plot(y1,x,color='blue')
plt.title('Plot3')

plt.subplot(2,2,4)
plt.plot(y2,x,color='yellow')
plt.title('Plot4')
```

```python
## Bar Graph

categories = ['A','B','C','D','E']

values = [5,7,3,8,6]

## Create Bar plot

plt.bar(categories,values,color='blue')
```

```python
## Histogram (used to show distribution of data)

data = [1,2,3,3,4,5,5,5,5,4,4,4,5,5]
## Create histogram
plt.hist(data,bins=5,color='orange',edgecolor='black')
```

```python
## Creating a scatter Plot

x= [1,2,3,4,5]
y = [2,3,4,5,6]

plt.scatter(x,y,color='purple',marker='x')
```

```python
## pie chart

label = ['A','B','C','D']
size = [30,20,40,10]
colors = ['gold','yellowgreen','lightcoral','lightskyblue'] 
explode = (0.1,0,0,0) ## This will move out the first slice

## create pie chart
plt.pie(size,labels=label,explode=explode,colors=colors,autopct = "%1.1f%%",shadow= True)
```

```python
import pandas as pd
df= pd.read_csv('sales_data.csv')
df.info()
```

```python
total_sales_by_product = df.groupby('Product Category')['Total Revenue'].sum()
print(total_sales_by_product)
```

```python
total_sales_by_product.plot(kind= 'bar',color='skyblue')
```

```python
## Plot sales trend over time
df1 = df.groupby('Date')['Total Revenue'].sum().reset_index()
df1
plt.plot(df1['Date'],df1['Total Revenue'])
```
